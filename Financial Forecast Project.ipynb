# Importar Librerías

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import root_mean_squared_error
```

# Cargar el Dataset

```python
file_path = "C:/Users/affec/Financial Forecast Dataset/RoS_train.csv"
df = pd.read_csv(file_path)
```

# EDA

```python
print(df.info())
print(df.describe())
```

# Visualización de la distribución de ventas

```python
sales_scaler = MinMaxScaler()
sales_scaler.fit(df[["Sales"]])
df["Sales"] = sales_scaler.inverse_transform(df[["Sales"]])

plt.figure(figsize=(10, 6))
sns.histplot(df["Sales"], bins=50, kde=True, color="blue")
plt.title("Distribución de Ventas")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

# Limpieza y preprocesamiento de datos

```python
# Imputar valores nulos
df = df.assign(CompetitionDistance=df["CompetitionDistance"].fillna(df["CompetitionDistance"].median()))
df.fillna(0, inplace=True)

# Convertir variables categóricas a numéricas
df["StateHoliday"] = df["StateHoliday"].astype(str)
df["StoreType"] = df["StoreType"].astype("category").cat.codes
df["Assortment"] = df["Assortment"].astype("category").cat.codes
```

# Feature Engineering

```python
df["SalesPerCustomer"] = df["Sales"] / df["Customers"]
df["YearsSinceCompetition"] = df["year"] - df["CompetitionOpenSinceYear"]
df["YearsSinceCompetition"] = df["YearsSinceCompetition"].apply(lambda x: x if x > 0 else 0)
df["PromoActive"] = df["Promo"].apply(lambda x: 1 if x > 0 else 0)

# Normalización de datos
scaler = MinMaxScaler()
df[["Sales", "Customers", "SalesPerCustomer"]] = scaler.fit_transform(df[["Sales", "Customers", "SalesPerCustomer"]])
```

# División de datos

```python
features = ["Customers", "Open", "PromoActive", "StoreType", "Assortment", "CompetitionDistance", "YearsSinceCompetition", "SalesPerCustomer"]
target = "Sales"

X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)
```

# Modelado 

```python
## 📈 Regresión Lineal
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
y_pred_lr = lr_model.predict(X_test)
rmse_lr = root_mean_squared_error(y_test, y_pred_lr)
r2_lr = r2_score(y_test, y_pred_lr)

## 🌲 Random Forest con optimización de memoria
rf_model = RandomForestRegressor(n_estimators=50, max_depth=10, random_state=42)
rf_model.fit(X_train.sample(frac=0.3, random_state=42), y_train.sample(frac=0.3, random_state=42))
y_pred_rf = rf_model.predict(X_test)
rmse_rf = root_mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
```

# Evaluación de modelos

```python
print(f"📊 Regresión Lineal -> RMSE: {rmse_lr:.4f}, R2: {r2_lr:.4f}")
print(f"🌲 Random Forest -> RMSE: {rmse_rf:.4f}, R2: {r2_rf:.4f}")
```

```python
model_results = pd.DataFrame({
    "Modelo": ["Regresión Lineal", "Random Forest"],
    "RMSE": [rmse_lr, rmse_rf],
    "R²": [r2_lr, r2_rf]
})

plt.figure(figsize=(8, 5))
sns.barplot(x="Modelo", y="RMSE", data=model_results, hue="Modelo", palette="viridis", legend=False)
plt.title("Comparación de Error (RMSE) entre Modelos")
plt.ylabel("RMSE")
plt.show()
```

## 📌 Conclusiones y Recomendaciones
- El modelo de **Random Forest** predice ventas con un **R² del 99.8%**, lo que indica una alta precisión.
- Las variables más influyentes en las ventas son:
  
  1️⃣ Número de clientes.

  2️⃣ Promociones activas.

  3️⃣ Distancia de la competencia.
